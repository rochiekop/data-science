
Lakukan imputasi nilai median pada kolom `gross`, `budget`, dan `num_critic_for_reviews` dan  nilai modus pada kolom `title_year`

**NOTE**

`The following can be seen as standard guideline for using mean, median or mode for replacing the missing values:`


*   `Mean` imputation is often used when the missing values are numerical and the distribution of the variable is approximately normal.
*   `Median` imputation is preferred when the distribution is skewed, as the median is less sensitive to outliers than the mean.
*   `Mode` imputation is suitable for categorical variables or numerical variables with a small number of unique values.






```python
# impute gross, budget, dan num_critic_for_reviews dengan median
movies['gross'] = movies['gross'].fillna(movies['gross'].median())
movies['budget'] = movies['budget'].fillna(movies['budget'].median())
movies['num_critic_for_reviews'] = movies['num_critic_for_reviews'].fillna(movies['num_critic_for_reviews'].median())
```

```python
# impute gross, budget, dan num_critic_for_reviews dengan mode
movies['title_year'] = movies['title_year'].fillna(movies['title_year'].mode()[0])
```

```python
movies.isna().sum()
```

Pada kolom `num_user_for_reviews` karena kolom tersebut merepresentasikan data numeric namun bertipe object, maka akan dilakukan inspeksi terlebih dahulu

```python
movies['num_user_for_reviews'].unique()
```

```python
# mengganti nilai nan dan string kosong dengan nilai 0
movies["num_user_for_reviews"] = movies["num_user_for_reviews"].fillna(0)
movies["num_user_for_reviews"] = movies["num_user_for_reviews"].replace({" ":0})
```

```python
movies['num_user_for_reviews'].unique()
```

Agar dapat diimpute dengan nilai statistik yang representatif, akan maka dilakukan casting pada kolom `num_user_for_reviews` menjadi bertipe int64.

Note: data pada kolom `gross`, `budget`, `num_critic_for_reviews` dan `title_year` juga akan dicasting menjadi int64 karena datanya dapat direpresentasikan dalam bentuk int.

```python
movies['num_user_for_reviews'] = movies['num_user_for_reviews'].astype('int64')
movies['gross'] = movies['gross'].astype('int64')
movies['title_year'] = movies['title_year'].astype('int64')
movies['num_critic_for_reviews'] = movies['num_critic_for_reviews'].astype('int64')
```

```python
movies.info()
```

```python
sns.histplot(data=movies, x="num_user_for_reviews")
```

```python
# Mengganti nilai 0 dengan nilai median dari data
movies["num_user_for_reviews"] = movies["num_user_for_reviews"].replace({0: movies["num_user_for_reviews"].median()})
```

```python
# mengganti nilai nan dengan unknown
movies["director_name"] = movies["director_name"].fillna("unknown")
movies["language"] = movies["language"].fillna("unknown")
movies["actor_1_name"] = movies["actor_1_name"].fillna("unknown")
```
