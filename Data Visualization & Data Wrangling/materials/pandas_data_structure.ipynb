## **Study Case 1: Mengolah nilai mahasiswa**
___

Anda mendapat tugas untuk mengolah data nilai mahasiswa di bawah ini.

|Nama|Nilai|
|-|-|
|Juchiro|83.54|
|Fandy|86.79|
|Erlangga|88.57|
|Matheos|71.48|
|Sugeng|72.65|
|Atma|92.73|
|Muliana|83.22|
|Iyya|77.72|
|Budiarto|85.09|
|Afendy|88.98|
|Muhaimin|68.40|

### 1.1
Buat series dari nilai tersebut. Nama sebagai index, nilai sebagai value


```python
# buat data nilai dalam bentuk python dictionary
data_nilai = {
    "Juchiro": 83.54,
    "Fandy":86.79,
    "Erlangga":88.57,
    "Matheos":71.48,
    "Sugeng":72.65,
    "Atma":92.73,
    "Muliana":83.22,
    "Iyya":77.72,
    "Budiarto":85.09,
    "Afendy":88.98,
    "Muhaimin":68.40,
}

# buat series dari data dictionary
data = pd.Series(data_nilai)

# tampilkan data
data
```

### 1.2
Tampilkan nilai Fandy dan Atma secara terpisah. Serta Tampilkan nilai Erlangga, Budiarto, dan Muhaimin bersamaan.

```python
print('Nilai Fandy : ',data['Fandy'])
print('Nilai Atma : ',data['Atma'])
data[["Erlangga", "Budiarto", "Muhaimin"]]
```

### 1.3
Tambahkan data berikut pada series yang telah dibuat

|Nama|Nilai|
|-|-|
|Yusril|86.35|
|Priyo|74.73|
|Yosua|79.12|

```python
new_data = {
    "Yusril" : 86.35,
    "Priyo"  : 74.73,
    "Yosua"  : 79.12
}
n_series = pd.Series(new_data)
n_series
```

```python
data = data.append(n_series)
data
```

### 1.4
Ubah nilai mahasiswa berikut ini:
* Muliana menjadi 92.45
* Iyya menjadi 79.80
* Juchiro dan Fandi masing-masing menjadi 90 dan 91 (secara bersamaan)

```python
data["Muliana"]
```

```python
data[["Muliana","Iyya","Juchiro","Fandy"]] = [92.45,79.80,90,91]
```

```python
data
```

### 1.5
Export data menjadi file csv dengan nama "nilai_mahasiswa_update.csv"


```python
data.to_csv("nilai_mahasiswa_update.csv")
```

___

Anda mendapat tugas sebagai admin dalam sebuah mall. Anda mendapat tugas untuk melakukan beberapa perubahan pada data pelanggan mall tersebut. Tapi, data tersebut tersimpan dalam bentuk file csv dengan nama "Mall_Customer.csv".

### 2.1
Buka data file csv tersebut dalam bentuk dataframe.

```python
# Baca data csv menggunakan fungsi pada pandas.
data = pd.read_csv("Mall_Customers.csv")

# tampilkan data
data
```

```python
data.shape
```

### 2.2
Index pada data merupakan angka dari 0 hingga 199. Anda diminta untuk mengubah Index menjadi angka hasil penjumlahan 1000 dengan "CustomerID". Misalnya CustomerID = 9, index menjadi 1009. Kemudian hapus  kolom "CustomerID"

```python
idx = data["CustomerID"].values + 1000
data.index = idx
data
```

```python
data
```

```python
# Drop CustomerId
data.drop("CustomerID",axis=1, inplace=True)
data
```

### 2.3
Customer 1098 dan 1099 tidak lagi menjadi pelanggan pada mall tersebut.Anda diminta untuk menghapus data index 1098 dan 1099.

```python
data = data.drop([1098,1099],axis=0)
data
```

### 2.4 
Ada data baru pada data customer. Anda Harus menambahkan data baru berikut.

||Genre|Age|Annual_Income_(k$)|Spending_Score
|-|-|-|-|-|
|1201|Male|22|20|83

```python
n_df = pd.DataFrame({
    "Genre" : "Male",
    "Age"   : 22,
    "Annual_Income_(k$)":20,
    "Spending_Score":83},
    index=[1201]
)
n_df
```

```python
data = data.append(n_df)
data
```

### 2.5
Data income dan score kedepannya berpotensi berisi data float. Ubah type data income dan score menjadi float.

```python
data['Annual_Income_(k$)'] = data['Annual_Income_(k$)'].astype(float)
data['Spending_Score'] = data["Spending_Score"].astype(float)
data
```

```python
# tampilkan type data tiap kolom
data.dtypes
```

## Study Case 3: Menggabungkan data pasien rumah sakit dari file berbeda.
Ada 4 file dari 2 rumah sakit yang berbeda yang harus anda gabung.
* File "patient_A1.csv" merupakan data pasien dari rumah sakit A yang berisi data Pregnancies, Glucose, BloodPressure, SkinThickness.
* File "patient_A2.csv" merupakan data pasien dari rumah sakit A yang berisi data Insulin, BMI, Diabetes, PedigreeFunction, Age, Outcome.
* File "patient.xlsx" merupakan data pasien dari rumah sakit B yang berisi data 
  - sheet patient_B1: IdPatient, Pregnancies, SkinThickness, BMI, Age, Outcome.
  - sheet patient_B2: IdPatient, Glucose, BloodPressure, DiabetesPedigreeFunction.

### 3.1
Buat dataframe dari setiap file csv yang diberikan.

```python
# Baca data csv menggunakan fungsi pada pandas.
dataA1 = pd.read_csv("patient_A1.csv")

# tampilkan data
dataA1.head()
```

```python
dataA2 = pd.read_csv("patient_A2.csv")
dataA2.head()
```

```python
dataB1 = pd.read_excel("patient.xlsx",sheet_name="patient_B1")
dataB1.tail()
```

```python
dataB2 = pd.read_excel("patient.xlsx",sheet_name="patient_B2")
dataB2.tail()
```

### 3.2
Anda harus menggabungkan 4 file tersebut menjadi 1 dataframe.


```python
dataA = pd.concat([dataA1,dataA2],axis=1)
dataA
```

```python
dataB = dataB1.merge(dataB2, on="IdPatient")
dataB
```

```python
# Drop IdPatient
dataB = dataB.drop("IdPatient", axis= 1)
dataB
```

```python
# gabungkan data rumah sakit A dan rumah sakit B
data = pd.concat([dataA, dataB], ignore_index=True)

# tampilkan data
data
```

### 3.3
Anda diminta menghapus data Pregnancies, DiabetesPedigreeFunction, dan Outcome karena tidak dibutuhkan


```python
#  hapus data Pregnancies, DiabetesPedigreeFunction, dan Outcome
data = data.drop(["Pregnancies","DiabetesPedigreeFunction","Outcome"], axis=1)
data
```
