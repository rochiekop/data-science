### **1. Looping in Python**
`Looping` atau **perulangan** adalah metode di dalam sebuah bahasa pemrograman untuk melakukan suatu pekerjaan seperti menampilkan data, memproses data atau pekerjaan yang lainnya yang sifatnya **repetitif** atau **berulang** (**loop**). Misalnya seperti:
 - Menambahkan angka 100 kali
 - Menulis kalimat yang sama sebanyak 50 kali
 - Mencari channel TV yang bagus
 - Fotocopy sebuah KTP sebanyak 5 lembar
 - dll

Pada bahasa pemrograman Python ada **dua** jenis `looping`, yaitu
 - `for`
 - `while`

### **2. For Loop**

Struktur perulangan for adalah sebagai berikut

```python
for itarator_variable in sequence_name:
	Statements
	. . .
	Statements
```

Kita bisa melakukan `for` loop pada tipe data lain seperti angka, string dan `Composite Data Structure`.

**TAMBAHAN**

Pada loop kita bisa menggunakan `range` dan `len` secara bersamaan. `range` akan membuat kumpulan bilangan sesuai dengan angka yang dikembalikan oleh fungsi len. Dan `len` akan mengembalikan panjang dari suatu variabel .

Contohnya: Kita akan print element suatu list

### **3. While Loop**

Struktur perulangan pada while adalah sebagai berikut
```python
while <condition>:
    # Jika kondisi benar
    run_this_statement
```
Pada while kita bisa menggunakan while-else untuk menjalankan satu blok kode lain ketika kondisi pada while sudah tidak terpenuhi

### **4. Combine Looping with Branching**

Looping bisa dikombinasikan dengan branching `(if-elif-else)` untuk menangani kondisi tertentu.

Misalkan seperti memeriksa sebuah angka ganjil atau genap, atau jika dalam kehidupan sehari-hari bisa digunakan saat mencari channel tv yang bagus dan lain-lain

### **5. Loop Control Flow**
Pada saat melakukan loop atau perulangan ada saatnya kita ingin keluar langsung dari perulangan tersebut pada suatu kondisi. Hal ini dapat dilakukan dengan yang namanya `Loop Control Flow`.
- Break
 - break akan **memaksa** perulangan untuk **berhenti** jika menemui suatu kondisi
- Continue
 - Jika `break` memaksa iterasi untuk berhenti maka `continue` akan memaksa untuk **eksekusi iterasi berikutnya**.
- Pass
 - Berbeda dengan `break` dan `continue` yang akan membuat perbedaan pada proses eksekusi program, `pass` **tidak melakukan apapun**. `pass` dalam Python digunakan ketika pernyataan diperlukan secara sintaksis tetapi kita tidak ingin perintah atau kode apa pun untuk dieksekusi.

### **6. List Comprehension**
Dengan List Comprehension kita bisa menyederhanakan operasi loop pada sebuah list

Loop Comprehension juga mendukung nested-for dan branching

## **Case Study 2: TV Quiz**
___


Sebuah acara di televisi yang diikuti oleh 200 orang sedang membuat quiz dan akan mengambil 3 volunteer dari penonton. TV tersebut membuat cara perhitungan yang unik untuk mengambil penonton secara acak

### 2A.  


Untuk volunteer pertama ambil penonton dengan ketentuan nomor urut yang bisa dibagi 7 dan merupakan hasil perkalian angka 5. Jika ada lebih satu bilangan ambil bilangan dengan urutan **ke 2 dari belakang**

````python
# buatlah sebuah empty list untuk menyimpan seluruh penonton yang terpilih
viewers = []

# buatlah variabel konstant untuk menyimpan jumlah penonton
NUMBER_VIEWER = 200

# buatlah variabel yang digunakan untuk menghentikan kondisi while
...
dived_by = 7
multiplication_by = 5
# buatlah empty list untuk menyimpan penonton terpilih pada persyaratan pertama
...
selected_viewer = []
# cari penonton dengan perulangan while sesuai kondisi
viewer = 1
while viewer <= NUMBER_VIEWER:
    if viewer % 7== 0 and viewer %5==0:
        viewers.append(viewer)
    viewer+=1

# print penonton yang memenuhi persyaratan
print(f"Penonton yang memenuhi persyaratan adalah nomor {viewers}")
# karena ada lebih dari satu angka maka ambil sesuai kondisi
# masukkan penonton terpilih ke variabel untuk menyimpan penonton yang terpilih
# hint (gunakan method append)
selected_viewer.append(viewers[-2])

# print penonton terpilih
print(f"Penonton Terpilih pertama adalah penonton dengan nomor urut : {selected_viewer[0]}")
````

### 2B.  

Untuk volunteer ke dua adalah penonton dengan nomor urut yang angkanya kurang dari 50 dan merupakan hasil perkalian dari angka 11. Jika ada lebih dari satu angka ambil bilangan **ke dua dari depan**.

````python
# buatlah variabel yang digunakan untuk menghentikan kondisi while
angka=1

# buatlah empty list untuk menyimpan penonton terpilih pada persyaratan kedua
penonton_terpilih_2 = []

# cari penonton dengan perulangan while sesuai kondisi
while angka <= NUMBER_VIEWER:
    if angka < 50 and angka % 11 == 0:
        penonton_terpilih_2.append(angka)
    angka += 1

# print penonton yang memenuhi persyaratan
print(f'Penonton yang memenuhi persyaratan adalah nomor {penonton_terpilih_2}')

# karena ada lebih dari satu angka maka ambil sesuai kondisi
# update variabel penonton yang terpilih dengan append
selected_viewer.append(penonton_terpilih_2[2])

# print penonton terpilih
print(f'Penonton Terpilih pada kondisi kedua adalah penonton dengan nomor urut : {penonton_terpilih_2[2]}')
````

### **2C**.

Untuk volunteer ke tiga adalah penonton yang memiliki angka total penjumlahan dari angka kelipatan 8 antara 50 sampai 100. Jika hasilnya lebih dari satu angka, ambil **angka yang paling belakang**



````python
# buatlah variabel yang digunakan untuk menghentikan kondisi while
viewer = 1

# buatlah empty list untuk menyimpan penonton terpilih pada persyaratan ketiga
penonton_terpilih_3 = []

# cari penonton dengan perulangan while sesuai kondisi
while viewer <= NUMBER_VIEWER:
    if viewer >= 50 and viewer <= 100:
        if viewer % 8 == 0:
            penonton_terpilih_3.append(viewer)
    viewer +=1
# print penonton yang memenuhi persyaratan
print(f"Penonton yang memenuhi persyaratan adalah nomor {penonton_terpilih_3}")

# karena ada lebih dari satu angka maka ambil sesuai kondisi
# update variabel penonton yang terpilih dengan append
selected_viewer.append(penonton_terpilih_3[-1])

# print penonton terpilih
print(f"Penonton Terpilih ketiga adalah penonton dengan nomor urut : {penonton_terpilih_3[-1]}")
````

## **Case Study 3: Total Belanja**
___


Bu Surti akan berbelanja di swalayan dengan Bu Mima dan Bu Rani


```
belanjaan ibu surti = 20.000, 90.000, 40.000, 160.500
belanjaan ibu mima = 31.000, 43.000, 65.000, 100.000
belanjaan ibu rani = 12.000, 200.000, 61.000, 10.000
```

### 3A.  


Buatlah list untuk menampung masing-masing harga dari belanjaan ibu-ibu tersebut.

````python
# buatlah list untuk masing-masing belanjaan
surti = [20_000, 90_000, 40_000, 160.500]
mima  = [31_000, 43_000, 65_000, 100_000]
rani  = [12_000, 200_000, 61_000, 10_000]
````

### 3B.

Hitunglah dengan menggunakan while loop total belanjaan mereka.

**Belanjaan Ibu Surti**

````python
# Gunakan variabel dari soal 3A

# buatlah variabel untuk menghentikan loop dahulu
length_s = len(surti)

# buat juga variabel untuk menampung total belanjaan ibu surti
total_payment_s = 0

# hitung belanjaan ibu surti dengan while loop
idx = 0
while idx < length_s:
    total_payment_s = surti[idx] + total_payment_s
    idx +=1
# print total belanja ibu surti
print(f"Total belanja ibu Surti = Rp {total_payment_s}")
````

**Belanjaan Ibu Mima**

````python
# buatlah variabel untuk menghentikan loop dahulu
length_m = len(mima)

# buat juga variabel untuk menampung total belanjaan ibu surti
total_payment_m = 0

# hitung belanjaan ibu mima dengan while loop
idx = 0
while idx < length_m:
    total_payment_m = mima[idx] + total_payment_m
    idx +=1
# print total belanja ibu mima
print(f"Total belanja ibu Mima = Rp {total_payment_m}")
````

**Belanjaan Ibu Rani**

````python
# buatlah variabel untuk menghentikan loop dahulu
length_r = len(rani)

# buat juga variabel untuk menampung total belanjaan ibu surti
total_payment_r = 0

# hitung belanjaan ibu rani dengan while loop
idx = 0
while idx < length_r:
    total_payment_r = rani[idx] + total_payment_r
    idx +=1
# print total belanja ibu rani
print(f"Total belanja ibu Rani = Rp {total_payment_r}")
````

````python
# definisikan potongan harga sebagai variabel konstan
DISCOUNT_1 = 1-0.01
DISCOUNT_2 = 1-0.03
````

````python
# Karena ambulance dan pemadan dikecualikan
# simpan ambulance dan pemadam kebakaran pada suatu tuple
...

# buat perulangan dengan kondisi
# cek apakah plat tersebut ganjil atau genap dan
# pastikan lewat jalur yang benar
# jika tidak lewat jalur yang benar proses tilang

for plat, jenis_kendaraan, jalur in data_kendaraan:
  
  temp = plat[-4]
  print(plat)    
  print(plat[-4])

  if int(temp) %2 != 0 and jalur == 'Jalur A': # Ganjil
    continue
  elif int(temp) %2 == 0 and jalur == 'Jalur B': # Genap
    continue
  else:
    print(f'{jenis_kendaraan} Plat Nomor {plat} masuk jalur {jalur} | Proses e-Tilang')
````
