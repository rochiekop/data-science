# Task 1. Calculate Average Sessions

**Task Descriptions**

- PacBook is a Social Media platform and they want to know what is their user Average Sessions and Session Duration for each data
- Given a three input for `load_time` (in minutes) and `exit_time` (in minutes)

- To calculate Session Duration, you can use this formula

$$
\text{Session}_\text{duration} = \text{exit}_\text{time} - \text{load}_\text{time}
$$

- To calculate Average Sessions, you can use this formula

$$
\text{Average}_\text{sessions} = \frac{\sum_{n=1}^{3}{\text{Session_duration}_{n}}}{\text{N_SESSIONS}}
$$

- But there are some constraints in this task
  - If the input for `load_time` and `exit_time` is less than zero or negative values, it will print the result **`"Not a Valid Input"`**

**Input - 1**

```py
load_time_1 = 5
exit_time_1 = 10

load_time_2 = 2
exit_time_2 = 15.5

load_time_3 = 1
exit_time_3 = 22.3
```

**Output - 1**

```py
Session 1 Duration (in minutes): 5
Session 2 Duration (in minutes): 13.5
Session 3 Duration (in minutes): 21.3
Average Session Duration (in minutes): 13.27
```

**Input - 2**

```py
load_time_1 = 0
exit_time_1 = 0

load_time_2 = 0
exit_time_2 = 0

load_time_3 = 0
exit_time_3 = 0
```

**Output - 2**

```py
Session 1 Duration (in minutes): 0
Session 2 Duration (in minutes): 0
Session 3 Duration (in minutes): 0
Average Session Duration (in minutes): 0
```

**Input - 3**

```py
load_time_1 = 3
exit_time_1 = 7

load_time_2 = -2
exit_time_2 = 5

load_time_3 = 3
exit_time_3 = 156
```

**Output - 3**

```py
Not a Valid Input
```

````python
# Write your code here
load_time_1 = 0
exit_time_1 = 0

load_time_2 = 0
exit_time_2 = 0

load_time_3 = 0
exit_time_3 = 0

# jawaban
n_session= 3 #total user
if (load_time_1 >= 0 and exit_time_1 >= 0) and (load_time_2 >= 0 and exit_time_2 >= 0) and (load_time_3 >= 0 and exit_time_3 >= 0):
  s_duration_1 = exit_time_1 - load_time_1
  s_duration_2 = exit_time_2 - load_time_2
  s_duration_3 = exit_time_3 - load_time_3
  avg_session = (s_duration_1+s_duration_2+s_duration_3)/n_session
  print(f"Session 1 Duration (in minutes): {s_duration_1}")
  print(f"Session 2 Duration (in minutes): {s_duration_2}")
  print(f"Session 3 Duration (in minutes): {s_duration_3}")
  print(f"Average Session Duration (in minutes): {0 if avg_session == 0 else round(avg_session,2)}")
else:
  print("Not a Valid Input")

````

````python
# Write your code here
temp = 13
from_unit = 'Kelvin'
to_unit = 'Celcius'
converted = temp
if from_unit.lower() == 'celcius' and to_unit.lower() != 'celcius':
  if to_unit.lower() == 'fahrenheit':
    converted = (temp * 9/5) + 32
  elif to_unit.lower() == 'kelvin':
    converted = temp + 273.15
elif from_unit.lower() == 'fahrenheit' and to_unit.lower() != 'fahrenheit':
  if to_unit.lower() == 'celcius':
    converted =  (temp-32) * 5/9
  elif to_unit.lower() == 'kelvin':
    converted = (temp-32) * 5/9 + 273.15
elif from_unit.lower() == 'kelvin' and to_unit.lower() != 'kelvin':
  if to_unit.lower() == 'celcius':
    converted =  temp- 273.15
  elif to_unit.lower() == 'fahrenheit':
    converted = (temp - 273.15)* 9/5 + 32

print(f"{temp} {from_unit} is equal to {converted} {to_unit}")
````

# Task 3. Extract NIK (Nomor Induk Kependudukan) Data

Category: Hard

---

**Task Descriptions**
- NIK is not some random numbers. You can extract information from the NIK.
  - The first six digits are a province, regency, and district code
  - The next six digits are the date of birth; DD-MM-YY, respectively
    - For men, the date of birth ranged between 01-31
    - For women, the date of birth is added with 40, hence ranged between 41-71.
  - The last 4 digits are a serial number of NIK issuance.


- Given **two** NIK, your tasks are
  1. Extract the date of birth of each NIK
  2. Extract their gender.
  3. Extract whether they have similar gender.
  4. Extract whether they have similar month of birth.

**Input**

```python
nik_1 = 1571011709860003
nik_2 = 2404075109910009
```

**Output**
```
Date of Birth
- NIK 1 : 17-09-86
- NIK 2 : 11-09-91

Gender
- NIK 1 : Male
- NIK 2 : Female

Do they have similar gender?
No

Do they have similar month of birth?
Yes, it is on September
```

**Questions 2**

---
Create your code based on the flowchart

````python
import calendar
# Write your code here
nik_1 = str(1571011709860003)
nik_2 = str(2404075109910009)

# extract the string number nik_1
date_1   = int(nik_1[6:8])
month_1  = nik_1[8:10]
year_1   = nik_1[10:12]
gender_1 = 'Male' if date_1 <=31 else "Female"
# extract the string number nik_2
date_2   = int(nik_2[6:8])
month_2  = nik_2[8:10]
year_2   = nik_2[10:12]
gender_2 = 'Male' if date_2 <=31 else "Female"

birth_date_1 = f"{date_1 if date_1 <=31 else date_1-40}-{month_1}-{year_1}"
birth_date_2 = f"{date_2 if date_2 <=31 else date_2-40}-{month_2}-{year_2}"


print("Date of Birth")
print("- NIK 1 :", birth_date_1)
print("- NIK 2 :", birth_date_2)
print()
print("Gender")
print("- NIK 1 :", gender_1)
print("- NIK 2 :", gender_2)
print()
print("Do they have similar gender?")
print(f'{"Yes" if gender_1==gender_2 else "No"}')
print()
print("Do they have similar month of birth?")
print(f'{"Yes, it is on {}".format(calendar.month_name[int(month_1)]) if int(month_1)==int(month_2) else "No"}')
````

# Task 4. Calculate Pajak Penghasilan (Simplified)

Category: Hard

---

**Task Descriptions**
- According to Pemotongan Pajak Penghasilan PPh Pasal 21 Tahun 2016, we can calculate our income tax as (we'll write in Bahasa Indonesia for avoiding missinterpretation)

- **Penghasilan tidak kena pajak (PTKP)**
  - Penghasilan Rp 54.000.000 setahun untuk wajib pajak.
  - Tambahan Rp 4.500.000 diberikan jika wajib pajak menikah.
  - Tambahan Rp 4.500.000 lagi jika punya 1 orang anak atau Rp 9.000.000 jika punya 2 orang anak. Apabila jumlah tanggungan lebih dari itu, maka tetap dihitung 3 orang tanggungan.

- **Tarif Pajak Penghasilan**
  - 5% untuk penghasilan kena pajak (PKP) sampai dengan 60 Juta
  - 15% untuk PKP di atas 60 juta sampai dengan 250 Juta
  - 25% untuk PKP di atas 250 juta sampai dengan 500 juta
  - 30% untuk PKP di atas 500 juta sampai dengan 5 miliar
  - 35% untuk PKP di atas 5 miliar

- Your task is calculate the income tax given
  - Monthly income
  - Marital status
  - The number of dependents (jumlah tanggungan)

**Input - 1**

```python
monthly_income = 4750000
marital_status = 'single'
n_dependents = 0
```

````python
monthly_income = 4750000
print(monthly_income * 12)
print((monthly_income * 12)*(5/100)-57000000)
````

````python
# Write your code here
monthly_income = 4750000
marital_status = 'single'
n_dependents = 0

# clasification PTKP
ptkp = 0
cat  = "-"
# constraint marital_satus 1.single/married
if marital_status.lower() == 'single':
  cat = "TK"
  if n_dependents == 0:
    ptkp = 54_000_000
  elif n_dependents == 1:
    ptkp = 58_500_000
  elif n_dependents == 2:
    ptkp = 63_000_000
  elif n_dependents >=3:
    ptkp = 67_500_000
elif marital_status.lower() == 'married':
  cat = "K"
  if n_dependents == 0:
    ptkp = 58_500_000
  elif n_dependents == 1:
    ptkp = 63_000_000
  elif n_dependents == 2:
    ptkp = 67_500_000
  elif n_dependents >=3:
    ptkp = 72_000_000
else:
  pass

nett_income    = monthly_income*12
taxable_income = nett_income - ptkp
# Clasification Tpp
pph_21 = 0
if taxable_income <= 60_000_000:
  pph_21 = (5/100)*taxable_income
elif taxable_income > 60_000_000 <= 250_000_000:
  pph_21 = (15/100)*taxable_income
elif taxable_income > 250_000_000 <= 500_000_000:
  pph_21 = (25/100)*taxable_income
elif taxable_income > 500_000_000 <= 5_000_000_000:
  pph_21 = (30/100)*taxable_income
else:
  pph_21 = (35/100)*taxable_income


print("Penghasilan neto setahun:")
print(nett_income)
print("\n")
print(f"PTKP {cat}/{n_dependents}")
print(-ptkp)
print("\n")
print("Penghasilan Kena Pajak:")
print(taxable_income)
print("\n")
print("PPh 21 terutang setahun:")
print(pph_21)
````
